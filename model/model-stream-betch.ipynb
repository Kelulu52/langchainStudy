# 1.流式调用（stream）

举例：

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="qwen3.6:27b",          # 只写模型名，与你在终端用 ollama run 的名字一致
    model_provider="ollama",
    base_url="http://10.20.62.52:11434"
)
```

```python
for response in model.stream("什么是agent?"):
    print(response.text,end="",flush=True)
```

# 2.批量调用（batch）性能优于invoke


举例：一次性调用多个问题

```python
messages = [
    "你是谁",
    "1+2=?",
    "中国首都叫什么"
]
response=model.batch(messages)
for response in response:
    print(response)
```

举例：按照完成顺序接收响应

```python
messages = [
    "你是谁",
    "1+2=?",
    "中国首都叫什么"
]
response=model.batch_as_completed(messages)
for response in response:
    print(response)
```
